```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Data
labels = [
    'CIDLog verifyLog',
    'SecLog verifyLog',
    'FT withdraw',
]
gas_costs = [
    688825,
    688825,
    696991,
]
colors = [
    'crimson',
    'forestgreen',
    'darkorange',
]

sorted_data = sorted(zip(gas_costs, labels, colors), key=lambda x: x[0])
sorted_gas, sorted_labels, sorted_colors = zip(*sorted_data)


plt.figure(figsize=(6.2, 3.8))  
bars = plt.bar(sorted_labels, sorted_gas, color=sorted_colors)


for bar, value in zip(bars, sorted_gas):
    plt.text(bar.get_x() + bar.get_width() / 2, value + 10, f'{value:,}', 
             ha='center', va='bottom', fontsize=8)

plt.ylabel('Gas', fontsize=10)
plt.title('Gas Consumption per Operation', fontsize=12)
plt.xticks(rotation=30, ha='right', fontsize=9)

legend_labels = [
    Patch(color='crimson', label='CIDLog verifyLog'),
    Patch(color='forestgreen', label='SecLog verifyLog'),
    Patch(color='darkorange', label='FT withdraw'),
]
plt.legend(handles=legend_labels, title="Operation", loc='center left',
           bbox_to_anchor=(1.0, 0.5), fontsize=8)

plt.tight_layout()
plt.savefig("SecLogGas.pdf", dpi=300, bbox_inches='tight')
plt.show()

```
