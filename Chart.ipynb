```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set()
  
# create data
x = ['100', '200', '300', '400', '500'] # number of transactions in each epoch
y5 = np.array([9202*100, 9202*200, 9202*300, 9202*400, 9202*500]) # each cost 9020 extra gas
y6 = np.array([21649*100, 21649*200, 21649*300, 21649*400, 21649*500]) # each cost 21649 extra gas
y7 = np.array([229777, 430656, 631186, 836101, 1042800]) # tested in truffle
y8 = np.array([20899*100, 20899*200, 20899*300, 20899*400, 20899*500]) # each cost 20899 extra gas
y9 = np.array([21000*10, 21000*20, 21000*30, 21000*40, 21000*50]) # each transfer transaction need 21000 gas
# each dataset is traded 10 times
# A standard ETH transfer requires a gas limit of 21,000 units of gas.
# https://ethereum.org/en/developers/docs/gas/#:~:text=An%20ETH%20transfer%20requires%2021%2C000,base%20fee%20is%2010%20gwei.

  
# plot bars in stack manner
plt.plot(x, y7+y9, label='Broker (Fair\u00b2Trade)') # total extra cost = cmt ver + coin_transfer * 10
plt.plot(x, y8, label='Buyer (Fair\u00b2Trade)')
plt.plot(x, y5, label='Broker (FairTrade*)')
plt.plot(x, y6, label='Buyer (FairTrade*)') # to be named as FairTrade* in the paper

plt.xlabel("Number of transactions in each epoch")
plt.ylabel("Gas")
plt.legend()
plt.title("Extra cost for revenue distribution")
#plt.savefig('TempTest.pdf', bbox_inches='tight')
plt.show()
```
